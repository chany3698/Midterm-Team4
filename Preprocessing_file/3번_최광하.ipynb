```python
# 2. 파일 불러오기 (한글 인코딩 대응: cp949 또는 euc-kr 사용)
import pandas as pd

filename = '서울상권분석.csv'  # 업로드한 파일명과 정확히 일치해야 함
df = pd.read_csv(filename, encoding='cp949')  # 또는 encoding='euc-kr'

# 3. 데이터 미리보기
df.head()
```

```python
# 결측치가 있는 열만 확인
missing = df.isnull().sum()
missing[missing > 0]
```

```python
# 원래 긴 이름의 컬럼을 간단하게 바꿔서 보기 편하게!
df.rename(columns={
    '서비스_업종_코드_명': '업종',
    '상권_코드_명': '상권명',
    '당월_매출_금액': '매출총액',
    '남성_매출_금액': '남성매출',
    '여성_매출_금액': '여성매출',
    '연령대_10_매출_금액': '매출_10대',
    '연령대_20_매출_금액': '매출_20대',
    '연령대_30_매출_금액': '매출_30대',
    '연령대_40_매출_금액': '매출_40대',
    '연령대_50_매출_금액': '매출_50대',
    '연령대_60_이상_매출_금액': '매출_60대이상',
}, inplace=True)

# 바뀐 컬럼 확인
df.columns
```

```python
#매출 합계 대비 성별, 연령대 비율 계산
# 성별 매출 비율
df['남성비율'] = df['남성매출'] / df['매출총액']
df['여성비율'] = df['여성매출'] / df['매출총액']

# 연령대별 비율
df['10대비율'] = df['매출_10대'] / df['매출총액']
df['20대비율'] = df['매출_20대'] / df['매출총액']
df['30대비율'] = df['매출_30대'] / df['매출총액']
df['40대비율'] = df['매출_40대'] / df['매출총액']
df['50대비율'] = df['매출_50대'] / df['매출총액']
df['60대비율'] = df['매출_60대이상'] / df['매출총액']
```

```python
#평일 VS 주말 매출 비교 컬럼 만들기
df['주중매출'] = df['주중_매출_금액']
df['주말매출'] = df['주말_매출_금액']
df['주말비율'] = df['주말매출'] / df['매출총액']
df['주중비율'] = df['주중매출'] / df['매출총액']
```

```python
#시간대 중 매출이 가장 높은 시간대 찾기
# 시간대 매출만 추출해서 가장 높은 시간대를 찾는 함수 만들기
def 최고매출시간대(row):
    시간대들 = {
        '00~06': row['시간대_00~06_매출_금액'],
        '06~11': row['시간대_06~11_매출_금액'],
        '11~14': row['시간대_11~14_매출_금액'],
        '14~17': row['시간대_14~17_매출_금액'],
        '17~21': row['시간대_17~21_매출_금액'],
        '21~24': row['시간대_21~24_매출_금액'],
    }
    return max(시간대들, key=시간대들.get)

df['최고매출시간대'] = df.apply(최고매출시간대, axis=1)
df.columns
```

```python
# 20대, 30대 매출 컬럼만 추출해서 새로운 데이터프레임 만들기
df_2030 = df[['상권명', '업종', '매출_20대', '매출_30대']]

# 미리보기
df_2030.head()
```

```python
# 다시 확인차 계산 (필요 시)
df_2030['매출_2030합'] = df_2030['매출_20대'] + df_2030['매출_30대']
```

```python
# 매출_2030합 기준으로 내림차순 정렬
df_2030_sorted = df_2030.sort_values(by='매출_2030합', ascending=False)

# 상위 10개만 미리보기
df_2030_sorted.head(10)
```

```python
df_2030_sorted.to_csv("상권_2030매출_정렬.csv", index=False)
```

```python
# 한글 깨짐 방지용으로 cp949 인코딩 사용
df_2030_sorted.to_csv("상권_2030매출_정렬_한글정상.csv", index=False, encoding='cp949')
```

```python
# '음식'이라는 단어가 업종명에 포함된 행만 추출
df_food = df_2030_sorted[df_2030_sorted['업종'].str.contains('음식', na=False)]
```

```python
# CSV 저장 (한글 인코딩 cp949)
df_food.to_csv("상권_2030매출_음식점_한글정상.csv", index=False, encoding='cp949')

# 다운로드
from google.colab import files
files.download("상권_2030매출_음식점_한글정상.csv")
```

```python
# 20대 매출과 30대 매출이 모두 0인 경우 삭제
df_food_clean = df_food[~((df_food['매출_20대'] == 0) & (df_food['매출_30대'] == 0))]

# 결과 미리보기
df_food_clean.head()
```

```python
# CSV로 저장 (한글 인코딩)
df_food_clean.to_csv("상권_2030매출_음식점_0제외.csv", index=False, encoding='cp949')

# 다운로드
from google.colab import files
files.download("상권_2030매출_음식점_0제외.csv")
```

```python
# 1. 20대 또는 30대 매출이 0인 음식점 제거
df_food_filtered = df_food[~((df_food['매출_20대'] == 0) | (df_food['매출_30대'] == 0))]

# 2. 업종명에 '양식' 포함된 것만 필터링
df_western = df_food_filtered[df_food_filtered['업종'].str.contains('양식', na=False)]

# 결과 미리보기
df_western.head()
```

```python
# 저장 (한글 인코딩)
df_western.to_csv("상권_2030매출_양식음식점_정제.csv", index=False, encoding='cp949')

# 다운로드
from google.colab import files
files.download("상권_2030매출_양식음식점_정제.csv")
```

```python
# 혹시 매출 합계 컬럼이 없다면 먼저 만들어줘
df_western['매출_2030합'] = df_western['매출_20대'] + df_western['매출_30대']

# 매출 합 기준으로 내림차순 정렬
df_western_sorted = df_western.sort_values(by='매출_2030합', ascending=False)

# 미리보기
df_western_sorted.head()
```
