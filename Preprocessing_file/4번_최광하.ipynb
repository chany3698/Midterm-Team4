```python
import pandas as pd

# 정확한 파일명 넣기 (업로드한 파일명!)
file_name = "장애인_기업 등록_현황.csv"

# 한글 인코딩 깨짐 방지를 위해 cp949 사용
df = pd.read_csv(file_name, encoding="cp949")

# 미리 보기
df.head()
```

```python
# 모든 기업이 장애인고용 기업이므로 1로 이진화
df['장애인고용'] = 1

# 결측치가 있는 행 모두 제거
df = df.dropna()

# 확인
df.info()
```

```python
# "소재지" 열에서 "포항"이라는 글자가 포함된 행만 선택
포항기업 = df[df['소재지'].str.contains('포항', na=False)]

# 결과 확인
포항기업.head()
```

```python
포항기업.to_csv("포항소재_장애인기업.csv", index=False, encoding='utf-8-sig')
from google.colab import files
files.download("포항소재_장애인기업.csv")
```
